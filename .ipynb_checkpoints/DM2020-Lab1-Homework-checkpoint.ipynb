## 2. Data Preparation

```python
# import the needed modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Dictionary with the files names and path to better reach them when needed
filepath_dict = {'yelp':   './sentiment labelled sentences/yelp_labelled.txt',
                 'amazon': './sentiment labelled sentences/amazon_cells_labelled.txt',
                 'imdb':   './sentiment labelled sentences/imdb_labelled.txt'}

# create a list of dataframes from the raw text data
df_list = []
for source, filepath in filepath_dict.items():
    df = pd.read_csv(filepath, names=['sentence', 'label'], sep='\t')
    df['source'] = source  # Add another column filled with the source name
    df_list.append(df)
    
# Now we have a list of dataframes, lets put them together as one dataframe
df = pd.concat(df_list)
```

```python
# Let's display a snapshot of the df
df.head()
```

```python
df.tail()
```

```python
# Let's see how big the dataset is
print("We have {} data points with {} features each".format(df.shape[0], df.shape[1]))
```

```python
# We can see from the tail that we only have 747 datapoints
# That's just a sampling problem because the dataset comes from different sources
# We can just reset the index as follows and remove the extra column
df = df.reset_index()
df = df.drop('index', 1)
```

```python
df.tail()
```

Everything is normal Now

Let's try to randomly print some texts from the dataframe. Fell free to run this cell many times to see different comments

```python
df.iloc[np.random.randint(0, 2748)][0]
```

Since we are here, let's add a new column label_string which contains "positive" if the label is 1 and negative if it's 0

```python
df['label_string'] = list(np.where(df.label == 1, 'positive', 'negative'))
```

```python
df.head()
```

Usually the labels (y) are placed at the leftmost column, let's do this.

```python
df = df.reindex(columns=['sentence','source', 'label_string', 'label'])
```

```python
df.head()
```

```python
# Check for missing values (we will need the helper)
from helpers import data_mining_helpers as dmh

df.isnull().apply(lambda x: dmh.check_missing_values(x))
```

```python
# Now let's check for duplicates
sum(df.duplicated())
```

Well, we have 17 of them, let's check them

```python
df[df.duplicated(keep=False)]
```

Now we should remove the duplicates, right?

```python
df.drop_duplicates(inplace=True, keep=False)
df = df.reset_index()
```

```python
# See how much data we have left now that the duplicates are removed
df.shape
```

```python
df.shape
```

```python
# but we don't really need the label_string, let's remove it
data = df.drop('label_string', 1)
```

```python
data['source'] = data['source'].cat.codes
```

## Part Three

### Visualizations

```python
# Bar chart to see the distribution of the data
# The data is quite balanced
print(data.label.value_counts())

# plot barchart for X_sample
data.label.value_counts().plot(kind = 'bar',
                                    title = 'Category distribution',      
                                    rot = 0, fontsize = 11, figsize = (8,3))
```

I love pie charts... They provide a clearer idea of the data distribution than bar charts

```python
# autopct create %, start angle represent starting point
data.label.value_counts().plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90° (
                            shadow=True,       # add shadow      
                            )

plt.title('Distribution of the labels between positive and negative')

plt.show()
```

Since we are dealing with word data, wouldn't it be nice to have a wordcloud ?

```python
# install wordcloud
!pip install wordcloud

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
# To remove redundant words like: I, is...
stopwords = set(STOPWORDS)
```

```python
# instantiate a word cloud object
review_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# Put all reviews into the same string and generate the word cloud
all_reviews_string=(" ").join(list(data['sentence']))
review_wc.generate(all_reviews_string)
```

Awesome! Now that the `word` cloud is created, let's visualize it.

```python
# display the word cloud
plt.imshow(review_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Interesting! So in the first 2000 words in the novel, the most common words are **good**, **movie**, **film**, and so on. Let's resize the cloud so that we can see the less frequent words a little better.

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(review_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Excellent! This looks really interesting! Another cool thing you can implement with the `word_cloud` package is superimposing the words onto a mask of any shape. Let's use a mask of baloon shape.

```python
from PIL import Image # converting images into arrays
# save mask to alice_mask
movie_mask = np.array(Image.open('image.jpg'))
```

Let's take a look at how the mask looks like.

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(movie_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# instantiate a word cloud object
review_wc = WordCloud(background_color='white', max_words=2000, mask=movie_mask, stopwords=stopwords)

# generate the word cloud
review_wc.generate(all_reviews_string)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(review_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
